```python
def burstPlot(x,y):
    c = (y >0)*(x>0)
    lx = np.log10(x[c])
    ly = np.log10(y[c])

    fit = S.linregress(lx,ly)
    print fit

    pl.plot(lx,ly,'.')
    pl.plot(lx,lx*fit[0]+fit[1],'k-')
    
    return fit
```
